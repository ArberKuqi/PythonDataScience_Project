#### ℹ️ Mbani Mend Metoda
##### `str.strip()` ne **Pandas🐼** përdoret për me hjek shkronjat e tepërta në fillim e në fund te (strings) - dmth me `str.strip()` i hek  hapsinat brenda nje string si Pershembull "Arber " me metoden `str.strip()` i hek hapsinat brenda nje string si Pershembull "Arber"

```python
ucl['Team'] = ucl['Team'].str.strip()
ucl['Competitions'] = ucl['Competitions'].str.strip()
```

```python
teams_per_comp = ucl.groupby(['Competitions'])['Team'].nunique().sort_values(ascending=False)
teams_per_comp
```
